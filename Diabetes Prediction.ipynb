```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/diabetes.csv")
df
```

```python
Y=df['Outcome']
Y
```

```python
X=df.drop(['Outcome'],axis='columns')
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=7)
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
Y_predicted=lr.predict(X_test)
for i in range(len(Y_predicted)):
  if Y_predicted[i]<0.5:
    Y_predicted[i]=0
  else:
    Y_predicted[i]=1
print("The First 5 Predicted Values are:")
print(Y_predicted[:5])
print("The First 5 Test Values are:")
print(np.array(Y_test[:5]))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,Y_train)
Y_predicted=knn.predict(X_test)
print("The First 5 Predicted Values are:")
print(Y_predicted[:5])
print("The First 5 Test Values are:")
print(np.array(Y_test[:5]))
knn.score(X_test,Y_test)
```

```python
from sklearn import metrics
matrix=metrics.confusion_matrix(Y_test,Y_predicted)
sns.heatmap(matrix,annot=True,fmt='0.1f',cmap='Blues')
plt.xlabel('Predicted Values')
plt.ylabel('Test Values')
plt.title('Confusion Matrix for KNeighbourClassifier')
plt.show()
```
